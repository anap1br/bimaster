```python
import pandas as pd
from selenium import webdriver
import time
```

```python
#driver = webdriver.Chrome()
driver = webdriver.Firefox()
```

**Parte 1 - Scraping dos 100 melhores filmes e séries de TV segundo a Empire**

```python
def scrap_empire():
    
    l_pos, l_img, l_title, l_date, l_review = [],[],[],[],[]

    # Para cada série da lista, guardar: título, data, review e endereço da imagem

    for num in range(0,100):
        img = driver.find_element_by_xpath("//div[@data-test='listicle-item-"+str(num)+"']//img").get_attribute("data-src")

        title = driver.find_element_by_xpath("//div[@data-test='listicle-item-"+str(num)+"']//h3").text
        i = title.find(')')
        if i==-1: i=-2
        title = title[i+2:]

        try:
            date = driver.find_element_by_xpath("//div[@data-test='listicle-item-"+str(num)+"']//strong").text
            if not date[:4].isdigit(): date='$'
        except:
            date='$'

        review = driver.find_element_by_xpath("//div[@data-test='listicle-item-"+str(num)+"']//p").text
        review = review.replace(date,'').replace("\n","")

        l_pos.append(100-num)
        l_img.append(img)
        l_title.append(title)
        l_date.append(date)
        l_review.append(review)
        
    dict_empire = {'#':l_pos,'Title':l_title,'Date':l_date,'Review':l_review,'Image':l_img}
        
    return dict_empire
```

Filmes

```python
%%time
url = r'https://www.empireonline.com/movies/features/best-movies-2/'
driver.get(url)
df_empire_movies = pd.DataFrame(scrap_empire())
df_empire_movies
```

```python
df_empire_movies.to_excel('empire_list_movies.xlsx')
```

Séries de TV

```python
%%time
url = r'https://www.empireonline.com/tv/features/best-tv-shows-ever-2/'
driver.get(url)
df_empire_tv = pd.DataFrame(scrap_empire())
df_empire_tv
```

```python
df_empire_tv.to_excel('empire_list_tv.xlsx')
```

**Parte 2 - Scraping no IMDB**

```python
url = r'https://www.imdb.com/'
driver.get(url)
```

```python
# Correção de alguns títulos
d_titles_tv = {'The Bridge':'Bron','Star Trek':'Star Trek Original', 'The Office (UK)':'The Office (2001)','Oz':'Oz (1997)', 'Angel':'Angel (1999)'}
d_titles_movies = {'Rocky':'Rocky (1976)','Predator':'Predator (1987)','Fargo':'Fargo (1996)'}
```

```python
def scrap_imdb (df_empire, tipo):
    l_title, l_grade, l_synopsis, l_awards, l_creator, l_actors, l_date, l_seasons, l_pop, l_length, l_budget, l_gross = [], [], [], [], [], [], [], [], [], [], [], []

    # Para cada série da lista da Empire, buscar o título no IMDB e salvar as informações

    for i in df_empire.index:

        title = df_empire.loc[i,'Title']
        
        try: 
            # Verifica se é preciso corrigir o título
            if tipo == 'tv':
                title = d_titles_tv[title]
            else: #movie
                title = d_titles_movies[title]
        except:
            pass
        
        try:
            # Busca o título e seleciona primeira opção
            driver.find_element_by_xpath("//input[@id='suggestion-search']").send_keys(title)
            driver.find_element_by_xpath("//button[@id='suggestion-search-button']").click()
            driver.find_element_by_xpath("//tr[@class='findResult odd'][1]/td[2]/a").click()

            # Grava as informações de interesse

            grade = driver.find_element_by_xpath("//span[contains(@class,'AggregateRatingButton__RatingScore')]").text

            popularity = driver.find_element_by_xpath("//div[contains(@class,'AggregateRatingButton__TotalRatingAmount')]").text
            if popularity[-1] == 'K':
                popularity = int(float(popularity[:-1])*1000)
            elif popularity[-1] == 'M':
                popularity = int(float(popularity[:-1])*1000000)

            synopsis = driver.find_element_by_xpath("//span[@data-testid='plot-l']").text
            if synopsis == '': synopsis = driver.find_element_by_xpath("//span[@data-testid='plot-xl']").text

            try:
                creator = driver.find_element_by_xpath("//li[@data-testid='title-pc-principal-credit'][1]/div/ul/li/a").text
            except:
                creator = ''

            try:
                awards = driver.find_element_by_xpath("//li[@data-testid='award_information']/a").text
                if awards == 'Awards': awards = driver.find_element_by_xpath("//li[@data-testid='award_information']/div/ul/li/span").text
            except:
                awards = ''
            
            c = '3' 
            if tipo == 'tv': c = '2' 
            actors = ''
            n = 1
            while n <= 3:
                try:
                    actors += ', ' + driver.find_element_by_xpath("//li[@data-testid='title-pc-principal-credit']["+c+"]/div/ul/li["+str(n)+"]/a").text
                    n += 1
                except:
                    break
            actors = actors[2:]

            if tipo == 'tv':
                date = driver.find_element_by_xpath("//ul[@data-testid='hero-title-block__metadata']/li[2]/a").text
                try:
                    seasons = driver.find_element_by_xpath("//select[@id='browse-episodes-season']/option[2]").text
                except:
                    seasons = '1'
                
                l_seasons.append(seasons)
                
            else: #movie
                date = driver.find_element_by_xpath("//ul[@data-testid='hero-title-block__metadata']/li[1]/a").text
                
                length = driver.find_element_by_xpath("//ul[@data-testid='hero-title-block__metadata']/li[3]").text
                
                budget = driver.find_element_by_xpath("//li[@data-testid='title-boxoffice-budget']/div/ul/li/span").text
                budget = budget.replace(',','').replace('$','').replace(' (estimated)','')
                try:
                    test = int(budget[:1])
                except: #se não for dólar, manter vazio
                    budget = ''
                    
                gross = driver.find_element_by_xpath("//li[@data-testid='title-boxoffice-cumulativeworldwidegross']/div/ul/li/span").text
                gross = gross.replace(',','').replace('$','').replace(' (estimated)','')
                
                l_length.append(length)
                l_budget.append(budget)
                l_gross.append(gross)

            l_title.append(df_empire.loc[i,'Title'])
            l_grade.append(grade)
            l_synopsis.append(synopsis)
            l_awards.append(awards)
            l_creator.append(creator)
            l_actors.append(actors)
            l_date.append(date)
            l_pop.append(popularity)
            
        except:
            print('Erro: '+ title)
        
    if tipo == 'tv': 
        dict_imdb = {'Title':l_title,'Date':l_date,'Grade':l_grade,'Grade_Count':l_pop,'Synopsis':l_synopsis,'Seasons':l_seasons,'Awards':l_awards,'Creator':l_creator,'Actors':l_actors}
    else: #movie
        dict_imdb = {'Title':l_title,'Date':l_date,'Grade':l_grade,'Grade_Count':l_pop,'Length':l_length,'Synopsis':l_synopsis,'Awards':l_awards,'Director':l_creator,'Actors':l_actors, 'Budget':l_budget,'Gross':l_gross}

    return dict_imdb
```

Filmes
